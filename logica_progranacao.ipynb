## Loops

```python
ages
```

```python
def driver(age):
    if age >= 18:
      print(f'{age} years old, can drive')
    else:
      print(f'{age} years old, cant drive')

for age in ages:
  driver(age)
```

```python
def driverList(ages):
  for age in ages:
    if age >= 18:
      print(f'{age} years old, can drive')
    else:
      print(f'{age} years old, cant drive')
```

```python
driverList(ages)
```

```python
permissions = []
```

```python
ages = [20, 14, 40]
```

```python
def permissionsDriver(ages, permissions):
  for age in ages:
    if age >= 18:
      permissions.append(True)
    else:
      permissions.append(False)
```

```python
permissionsDriver(ages, permissions)
```

```python
permissions
```

```python
for permission in permissions:
  if permission == True:
    print('Can drive')
  else:
    print('Cant drive')
```
